**FUNÇÃO USADA PARA DETERMINAR O MAIOR E O MENOR VALOR DE TODO O DATAFRAME PARA SER USADOS COMO VAI O QUANTO VAI VARIAR AS CORES DO MAPA"**

```
def minimo_maximo (df): #descobrir o menor e o maior valor valor da séries temporal para todas as federações
  lmin = list()
  lmax = list()

  for ano in range(2012,2017):
    l = list(df.loc[:,ano])
    l.pop(0)
    lmin.append(min(l))
    lmax.append(max(l))
  maximo = max(lmax)
  minimo = min(lmin)
  return [minimo,maximo]
```

**FUNÇÃO USADA CRIAÇÃO DOS MAPAS POR ANO USANDO A VARIAÇÃO DE COLOR BASEADO NO MÍNIMO E MÁXIMO DA DATAFRAME E USANDO O SHAPEFILE DO BRASIL ADQUIRIDO NA API DO IBGE"**

```
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection
import numpy as np
import matplotlib.colors

def gerar_maps (ano,df,nome,unidade): #descobrir o menor e o maior valor valor da séries temporal para todas as federações
  
  fig= plt.figure(figsize=(20,20))
  ax= fig.add_subplot(111)
  m = Basemap(llcrnrlon=-74.9,llcrnrlat=-34.5,urcrnrlon=-34.6,urcrnrlat=5.8,
               resolution='i', projection='merc')
 
  df_ano = df.loc[:,[ano]]
  df_ano.drop(index = ['1'], inplace =True) #Retirando dados nacionais do gráfico
  cmap= plt.cm.YlOrBr # ColorMap
  min_max = minimo_maximo(df)
  print(min_max)
  norm=plt.Normalize(min_max[0],min_max[1])
  m.readshapefile('malha_brasil-polygon', 'comarques',linewidth = 3) # lendo o shapefile do Brasil

  colvals = df_ano.loc[:][ano]
  patches = []
  for info, shape in zip(m.comarques_info, m.comarques):
      if info['codarea'] in df_ano.index:
        color = cmap(norm(df.loc[info['codarea']][ano]))
        patches.append( Polygon(np.array(shape), True, color=color) ) #Criando listas de cores do mapa

  pc = PatchCollection(patches, match_original=True, edgecolor='k', linewidths=1., zorder=2,) 
  ax.add_collection(pc)# colorindo o mapa
  plt.title('Mapa de '+nome+' '+str(ano),fontsize = 30)
  
  sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm) # criando o colorbar
  sm.set_array(colvals)
  cbar = fig.colorbar(sm, ax=ax) # colocando o colorbar
  cbar.ax.set_title(nome+unidade,fontsize = 20)
  plt.savefig('mapa'+str(ano)+nome+'.jpg',format ='jpg',quality = 100)
  
  return fig
```

**COM O AUXÍLIO DA BIBLIOTECA imageio JUNTAMOS TODAS OS MAPAS REFERENTE AS EMPRESAS E FAZENDO UM GIF"**

```
import imageio
def gerar_mapas_gifs(df,nome,unidade):
  imgs = []
  for ano in range(2012,2017): 
    imgs.append(gerar_maps(ano,df,nome,unidade))
  with imageio.get_writer('mapa_animado_'+nome+'.gif', mode='I',duration=0.5) as writer: #gerando gif
      for i in range(2012,2017):
          image = imageio.imread('mapa'+str(i)+nome+'.jpg')
          im = writer.append_data(image)
gerar_mapas_gifs(empresas,'empresas','')
gerar_mapas_gifs(desocupacao,'Desocupação','(Mil)')
```

**APÓS VISUALIZARMOS OS MAPAS ANTERIORES VAMOS VERIFICAR SE HÁ ALGUMA RELAÇÃO ENTRE AS VARIÁVEIS**

```
import scipy
import scipy.stats 
l1 = list()
for linha in desocupacao.iterrows():
  c,p= scipy.stats.pearsonr(desocupacao.loc[linha[0],[2012,2013,2014,2015,2016]],empresas.loc[linha[0],[2012,2013,2014,2015,2016]])
  l1.append([linha[0],c,c**2,p])
correlacao_df = pd.DataFrame(l1,columns=['codarea','Coeficiente','Coeficiente^2','Valor_P'])
correlacao_df.set_index('codarea',inplace =True)
correlacao_df = pd.merge(correlacao_df,empresas.loc[:,['codarea','federacao']],left_index = True,right_index =True).drop(columns = ['codarea'])
correlacao_df = correlacao_df[['federacao','Coeficiente','Coeficiente^2','Valor_P']]
correlacao_df.sort_values(by = 'Coeficiente')
```
